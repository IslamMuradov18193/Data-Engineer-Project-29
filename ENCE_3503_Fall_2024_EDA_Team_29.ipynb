```python
! pip install seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/scraped_data_team_29.csv')
df
```

```python
print(df.describe())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['rating'].fillna(df['rating'].mean(), inplace=True)
df['price_level'].fillna(df['price_level'].mean(), inplace=True)
df['phone_number'].fillna('Not Available', inplace=True)
df['website'].fillna('Not Available', inplace=True)
df.drop(columns=['opening_hours'], inplace=True)
```

```python
# Extract latitude and longitude from location
df['latitude'] = df['location'].str.extract(r"'lat': ([0-9.]+)").astype(float)
df['longitude'] = df['location'].str.extract(r"'lng': ([0-9.]+)").astype(float)

# Drop the original 'location' column
df.drop(columns=['location'], inplace=True)
```

```python
# Clean the 'types' column by removing square brackets and single quotes
df['types'] = df['types'].str.replace(r"[\\[\\]' ]", "", regex=True).str.split(",")
```

```python
# Transforming website to  has_website
df['has_website'] = df['website'].apply(lambda x: True if pd.notna(x) and x.strip() != "" else False)

# Drop website column
df.drop(columns=['website'], inplace=True)
# Display the updated DataFrame
print(df)
```

```python
# Clean the 'types' column 
df['types'] = df['types'].apply(lambda x: x[0]).str.replace('[',"")
```

```python
print(df.duplicated().sum())
print(df.isnull().sum())
print(df.info())
print(df.describe())
print(df['types'].unique())
```

```python
# Drop unneccessary columns 
df.drop(columns=['address','place_id','phone_number'], inplace=True)
```

```python
df.to_csv("data/processed_data_team_29.csv", index=False)
```

```python
# Use boxplots to identify outliers
sns.boxplot(x=df['rating'])
plt.title('Outliers in Ratings')
plt.show()
```

```python
q1 = np.percentile(df['rating'], 25)  # First quartile (25th percentile)
q2 = np.percentile(df['rating'], 50)  # Second quartile (50th percentile)
q3 = np.percentile(df['rating'], 75)  # Third quartile (75th percentile)

print(f"Q1: {q1}, Median: {q2}, Q3: {q3}")
```

```python
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Identify outliers
outliers = df[(df['rating'] < lower_bound) | (df['rating'] > upper_bound)]
print(outliers)

# Remove outliers (if needed)
df_cleaned = df[(df['rating'] >= lower_bound) & (df['rating'] <= upper_bound)]
```

```python
# Boxplot to show quartiles and outliers
sns.boxplot(x=df['rating'])
plt.title('Boxplot of Ratings with Quartiles and Outliers')
plt.show()
```

```python
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
